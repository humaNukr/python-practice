# I. Змінні та памʼять.

1. Створити змінні, що посилаються на два цілих числа, що однакові за значенням, де значення належить проміжку від -5 до 256. Перевірте, чи будуть ці змінні рівні тільки за значенням або ж ще будуть посилатися на один і той самий обʼєкт в памʼяті? Наведіть код та дайте текстову відповідь нижче.

```python
a = 100
b = 100

print(f"Value of a: {a}")
print(f"Value of b: {b}")

print(f"Are a and b equal by value? {a == b}")
print(f"Do a and b refer to the same object in memory? {a is b}")

a = -3
b = -3

print(f"Value of a: {a}")
print(f"Value of b: {b}")

print(f"Are a and b equal by value? {a == b}")
print(f"Do a and b refer to the same object in memory? {a is b}")

a = 257
b = 257

print(f"\nValue of a: {a}")
print(f"Value of b: {b}")

print(f"Are a and b equal by value? {a == b}")
print(f"Do a and b refer to the same object in memory? {a is b}")
```

Python виявився розумнішим, ніж я думав. Якщо числа маленькі, від -5 до 256, то він їх «запам'ятовує» і використовує той самий об'єкт, щоб не створювати новий. А от для великих чисел, як 257, він вже створює окремі об'єкти, навіть якщо вони однакові за значенням. Тому is для маленьких True, а для великих — False.

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set_1 = {24, 32, 42, 323}
set_2 = {321, 412, 24, 32}
print(f"Об'єднання: {set_1 | set_2}")
print(f"Різниця: {set_1 - set_2}")
print(f"Перетин: {set_1 & set_2} ")
print(f"Симетрична різниця: {set_1 ^ set_2}")
```

```python
my_dict = dict()

my_dict["name"] = "Artem"
my_dict[1] = "one"
my_dict[(2, 3)] = "tuple key"
my_dict[3.14] = "float key"

print(my_dict)
```
